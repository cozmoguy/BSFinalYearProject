```python
#importing essential libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#importing data from a csv file
stardata = pd.read_csv('stardata.csv')
```

```python
#droping null values
stardata = stardata.dropna()
```

```python
#shape of the data
stardata.shape
```

```python
#first 20 rows of data
stardata.head(20)
```

```python
#ploting graph using matplotlib library
x = stardata.iloc[:,8].values 
y = stardata.iloc[:,1].values
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_xlabel('B-V')
ax.set_ylabel('Vmag')
plt.scatter(x,y,s=1) # marker size
plt.ylim([y.min(axis=0),y.max(axis=0)]) # x-axis range
plt.xlim([x.min(axis=0),2]) # x-axis range
plt.gca().invert_yaxis()  # y-axis reversed
plt.figure(dpi=300)
plt.show()
```
